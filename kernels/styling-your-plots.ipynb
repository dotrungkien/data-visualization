```python
import pandas as pd
%matplotlib inline

reviews = pd.read_csv('../input/wine-reviews/winemag-data_first150k.csv', index_col=0)
reviews.head()
```

```python
reviews['points'].value_counts().sort_index().plot.bar()
```

```python
reviews['points'].value_counts().sort_index().plot.bar(figsize=(12, 6), color='mediumvioletred')
```

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6), 
    color='mediumvioletred', 
    fontsize=16,
    title='Rankings Given by Wine Magazine')
```

```python
import matplotlib.pyplot as plt
ax = reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6), 
    color='mediumvioletred', 
    fontsize=16)
print(type(ax))
ax.set_title('Rankings Given by Wine Magazine', fontsize=20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=15
)

ax.set_title("Rankings Given by Wine Magazine")
sns.despine(bottom=True, left=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pokemon = pd.read_csv('../input/pokemon/Pokemon.csv', index_col=0)
pokemon.head()
```

```python
pokemon.plot.scatter(x='Attack', y='Defense', title='Pokemon by Attack and Defense', figsize=(12,6))
```

```python
pokemon_by_stat_total = pokemon.groupby('Total')
ax = pokemon_by_stat_total.mean().plot.hist(figsize=(12, 6))
ax.set_title('Pokemon by Stat Total', fontsize=20)
```

```python
ax = pokemon['Total'].plot.hist(
    figsize=(12, 6),
    fontsize=15,
    bins=50,
    color='gray'
)

ax.set_title('Pokemon by Stat Total', fontsize=20)
```

```python
ax = pokemon['Type 1'].value_counts().plot.bar(
    figsize=(12, 6),
    fontsize=15
)

ax.set_title('Pokemon by Primary Type', fontsize=20)
sns.despine(left=True, bottom=True)
```
